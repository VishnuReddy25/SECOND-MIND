```python
# âœ… Generation Agent (Creates Hypothesis)
generation_agent = Agent(
    name="Generation Agent",
    role="Generate a research hypothesis based on search results.",
    # model=Gemini(id="gemini-2.0-flash-exp", api_key=api_key),
    model=model,
    instructions=[
    "1. Formulate a **well-defined, testable hypothesis** using search results.",
    "2. Use format: **'If [cause], then [effect], due to [reason].'**",
    "3. Keep responses **formal, clear, and scientifically valid.**",
    "4. **Do not repeat** the input question; reframe it as a hypothesis."
],
    show_tool_calls=True,
    markdown=True,
)

# âœ… Reflection Agent (Checks Hypothesis)
reflection_agent = Agent(
    name="Reflection Agent",
    role="Evaluate the quality and coherence of the hypothesis.",
    # model=Gemini(id="gemini-2.0-flash-exp", api_key=api_key),
    model=model,
    instructions=[
    "1. Verify if the hypothesis **directly answers the query.**",
    "2. Identify **gaps or logical flaws** in reasoning.",
    "3. Suggest refinements if it lacks specificity or clarity.",
    "4. Ensure it is **scientifically sound & well-structured.**"
],
    show_tool_calls=True,
    markdown=True,
)

# âœ… Ranking Agent (Scores Hypothesis)
ranking_agent = Agent(
    name="Ranking Agent",
    role="Score the hypothesis based on validity and relevance.",
    # model=Gemini(id="gemini-2.0-flash-exp", api_key=api_key),
    model=model,
    instructions=[
    "1. Score the hypothesis **on a scale of 1-10** based on:",
    "   - **Relevance**: Does it directly answer the query?",
    "   - **Scientific Validity**: Is it logical & research-backed?",
    "   - **Testability**: Can it be experimentally verified?",
    "   - **Clarity**: Is it structured and precise?",
    "2. Justify the score in **one short sentence**."
],
    show_tool_calls=True,
    markdown=True,
)

# âœ… Evolution Agent (Refines Hypothesis)
evolution_agent = Agent(
    name="Evolution Agent",
    role="Refine and improve the hypothesis based on feedback.",
    # model=Gemini(id="gemini-2.0-flash-exp", api_key=api_key),
    model=model,
    instructions=[
    "1. Improve the hypothesis **without changing its core meaning.**",
    "2. Add missing details while keeping it **concise and formal.**",
    "3. Ensure clarity, accuracy, and logical consistency.",
    "4. Integrate insights from past queries if relevant."
],
    show_tool_calls=True,
    markdown=True,
)

# âœ… Proximity Agent (Finds Related Research)
proximity_agent = Agent(
    name="Proximity Agent",
    role="Find related past research and relevant references.",
    # model=Gemini(id="gemini-2.0-flash-exp", api_key=api_key),
    model=model,
    instructions=[
    "1. Search for **previously stored related queries and hypotheses**.",
    "2. If relevant past data is found, **integrate insights into the response**.",
    "3. Ensure retrieved information **aligns with the current context.**",
    "4. If no past data exists, state that and proceed with new information.",
    "5. Avoid redundancyâ€”only include past data if it enhances the response."
],
    show_tool_calls=True,
    markdown=True,
)

# âœ… Meta-Review Agent (Final Quality Check)
meta_review_agent = Agent(
    name="Meta Review Agent",
    role="Perform a final review and suggest refinements.",
    # model=Gemini(id="gemini-2.0-flash-exp", api_key=api_key),
    model=model,
    instructions=[
    "1. Ensure the final output **directly addresses** the original query.",
    "2. Verify **clarity, structure, and logical flow.**",
    "3. Summarize key takeaways concisely.",
    "4. Suggest **one final minor improvement** if necessary."
],
    show_tool_calls=True,
    markdown=True,
)

# âœ… Supervisor (Manages Agent Workflow)
supervisor_agent = Agent(
    name="Supervisor Agent",
    role="Manages the workflow of all agents and ensures high-quality output.",
    # model=Gemini(id="gemini-2.0-flash-exp", api_key=api_key),
    model=model,
    team=[web_search_agent, generation_agent, reflection_agent, ranking_agent, evolution_agent, proximity_agent, meta_review_agent],
    instructions=[
        "Ensure smooth workflow where each agent builds on the previous one.",
        "Verify the final output before returning it to the user.",
    ],
    show_tool_calls=True,
    markdown=True,
)

# âœ… Function to Run Full Workflow
def execute_full_workflow(query):
    print("\nğŸ” **Step 1: Fetching Web Data...**")
    web_results = web_search_agent.run(
        input=query,
        expected_output="A summary of recent AI research on the topic.",
        additional_information="Extract key findings and avoid generic information.",
    )
    print("\nâœ… **Web Search Output:**\n", web_results)

    print("\nğŸ§  **Step 2: Generating Initial Hypothesis...**")
    hypothesis = generation_agent.run(
        input=web_results,
        expected_output="A structured research hypothesis.",
        additional_information="Ensure it's based on the summarized findings.",
    )
    print("\nâœ… **Generated Hypothesis:**\n", hypothesis)

    print("\nğŸ” **Step 3: Reviewing Hypothesis for Coherence...**")
    reflection_feedback = reflection_agent.run(
        input=hypothesis,
        expected_output="A review of hypothesis quality.",
        additional_information="Check for coherence and logical flaws.",
    )
    print("\nâœ… **Reflection Feedback:**\n", reflection_feedback)

    print("\nğŸ“Š **Step 4: Scoring Hypothesis...**")
    ranking_score = ranking_agent.run(
        input=hypothesis,
        expected_output="A score from 1-10 with justification.",
        additional_information="Rate based on accuracy, novelty, and credibility.",
    )
    print("\nâœ… **Hypothesis Score:**\n", ranking_score)

    print("\nğŸ”„ **Step 5: Refining Hypothesis...**")
    improved_hypothesis = evolution_agent.run(
        input=f"{hypothesis} \n\nReflection Feedback: {reflection_feedback}",
        expected_output="An improved version of the hypothesis.",
        additional_information="Incorporate suggested improvements.",
    )
    print("\nâœ… **Evolved Hypothesis:**\n", improved_hypothesis)

    print("\nğŸ“š **Step 6: Searching for Related Research...**")
    related_research = proximity_agent.run(
        input=improved_hypothesis,
        expected_output="Relevant past research references.",
        additional_information="Find similar studies or supporting data.",
    )
    print("\nâœ… **Related Research Found:**\n", related_research)

    print("\nğŸ›  **Step 7: Final Review and Refinement...**")
    final_review = meta_review_agent.run(
        input=improved_hypothesis,
        expected_output="A fully polished hypothesis ready for submission.",
        additional_information="Ensure clarity and proper formatting.",
    )
    print("\nğŸš€ **Final Processed Hypothesis:**\n", final_review)

    return final_review

# âœ… Run the full agent workflow
query = "How can AI help in climate change mitigation?"
final_output = execute_full_workflow(query)

print("\nğŸš€ **Final Processed Hypothesis:**\n", final_output)
```

```python
from typing import TypedDict
import re
from langgraph.graph import END, StateGraph, START
from langgraph.graph.state import CompiledStateGraph


class LLMAgentCall(TypedDict):
    query: str
    hypothesis: str
    feedback: str
    score: int
    refined_hypothesis: str
    proximity_score: float
    meta_review_text: str

# âœ… Supervisor Agent
def supervisor_agent(state: dict) -> dict:
    print(f"\n[SUPERVISOR] Deciding next step...")

    if "hypothesis" not in state:
        print("[SUPERVISOR] Sending to Generation...")
        return {"next": "generation"}

    elif "feedback" not in state:
        print("[SUPERVISOR] Sending to Reflection...")
        return {"next": "reflection"}

    elif "score" not in state:
        print("[SUPERVISOR] Sending to Ranking...")
        return {"next": "ranking"}

    elif "refined_hypothesis" not in state:
        print("[SUPERVISOR] Sending to Evolution...")
        return {"next": "evolution"}

    elif "proximity_score" not in state:
        print("[SUPERVISOR] Sending to Proximity Analysis...")
        return {"next": "proximity"}

    elif "meta_review_text" not in state:
        print("[SUPERVISOR] Sending to Meta Review...")
        return {"next": "meta_review"}

    else:
        print("[SUPERVISOR] Sending to Summary...")
        return {"next": "summary"}

# âœ… Generation: Creates initial hypotheses
def generative_agent(state: dict) -> dict:
    query = state["query"]
    hypothesis = f"Install solar panels on rooftops for urban renewable energy."

    print(f"[GENERATION] Hypothesis Created: {hypothesis}")
    return {"hypothesis": hypothesis}

# âœ… Reflection: Checks coherence
def reflection_agent(state: dict) -> dict:
    hypothesis = state["hypothesis"]
    feedback = f"Relevant but needs consideration for nighttime energy storage."

    print(f"[REFLECTION] Feedback Given: {feedback}")
    return {"feedback": feedback}

# âœ… Ranking: Scores hypothesis based on real-time data
def ranking_agent(state: dict) -> dict:
    score = random.randint(6, 10)  # Simulating ranking with external data

    print(f"[RANKING] Score Assigned: {score}")
    return {"score": score}

# âœ… Evolution: Refines hypothesis based on web search
def evolution_agent(state: dict) -> dict:
    refined_hypothesis = f"Use solar window panels instead of rooftops for urban areas."

    print(f"[EVOLUTION] Updated Hypothesis: {refined_hypothesis}")
    return {"refined_hypothesis": refined_hypothesis}

# âœ… Proximity: Links with past interactions
def proximity_agent(state: dict) -> dict:
    proximity_score = round(random.uniform(0.5, 1.0), 2)

    print(f"[PROXIMITY] Similarity Score: {proximity_score}")
    return {"proximity_score": proximity_score}

# âœ… Meta-review: Evaluates the process
def meta_review_agent(state: dict) -> dict:
    review = f"Web data caused Ranking delay; consider optimization."

    print(f"[META REVIEW] Review: {review}")
    return {"meta_review_text": review}

# âœ… Summary Agent
def summary_agent(state: dict) -> dict:
    summary = f"""
    Query: {state["query"]}
    Initial Hypothesis: {state["hypothesis"]}
    Reflection Feedback: {state["feedback"]}
    Initial Score: {state["score"]}
    Refined Hypothesis: {state["refined_hypothesis"]}
    Proximity Score: {state["proximity_score"]}
    Meta Review: {state["meta_review_text"]}
    """

    print("[FINAL SUMMARY]\n", summary)
    return {"summary": summary}

# âœ… Build Graph
builder = StateGraph(LLMAgentCall)

# âœ… Define nodes in correct order
builder.add_node("supervisor", supervisor_agent)
builder.add_node("generation", generative_agent)
builder.add_node("reflection", reflection_agent)
builder.add_node("ranking", ranking_agent)
builder.add_node("evolution", evolution_agent)
builder.add_node("proximity", proximity_agent)
builder.add_node("meta_review", meta_review_agent)
builder.add_node("summary", summary_agent)

# âœ… Define Supervisor as central router
builder.add_edge(START, "supervisor")
builder.add_conditional_edges("supervisor", lambda state: state)
builder.add_edge("generation", "supervisor")
builder.add_edge("reflection", "supervisor")
builder.add_edge("ranking", "supervisor")
builder.add_edge("evolution", "supervisor")
builder.add_edge("proximity", "supervisor")
builder.add_edge("meta_review", "supervisor")
builder.add_edge("summary", END)

# âœ… Compile Graph
graph = builder.compile()



```

```python
from IPython.display import Image, display
import time
import requests

def display_graph_with_retry(graph, max_retries=3, retry_delay=2):
    """Displays the graph visualization with retry mechanism."""
    for attempt in range(max_retries):
        try:
            # Removed timeout argument as it's not supported
            display(Image(graph.get_graph().draw_mermaid_png()))
            return  # Exit if successful
        except requests.exceptions.ReadTimeout as e:
            print(f"Attempt {attempt + 1} failed: {e}")
            if attempt < max_retries - 1:
                print(f"Retrying in {retry_delay} seconds...")
                time.sleep(retry_delay)
            else:
                print("Max retries reached. Giving up.")

# Call the function without timeout argument:
display_graph_with_retry(graph, max_retries=3, retry_delay=10)
```
