```python
from typing import TypedDict
import re
from langgraph.graph import END, StateGraph, START
from langgraph.graph.state import CompiledStateGraph

import random as random
class LLMAgentCall(TypedDict):
    query: str
    hypothesis: str
    feedback: str
    score: int
    refined_hypothesis: str
    proximity_score: float
    meta_review_text: str
    next: str



def supervisor_agent(state: dict) -> dict:
    print(f"\n[SUPERVISOR] Deciding next step...")

    if "hypothesis" not in state:
        print("[SUPERVISOR] Sending to Generation...")
        return {"next": "generation"}

    elif "feedback" not in state:
        print("[SUPERVISOR] Sending to Reflection...")
        return {"next": "reflection"}

    elif "score" not in state:
        print("[SUPERVISOR] Sending to Ranking...")
        return {"next": "ranking"}

    elif "refined_hypothesis" not in state:
        print("[SUPERVISOR] Sending to Evolution...")
        return {"next": "evolution"}

    elif "proximity_score" not in state:
        print("[SUPERVISOR] Sending to Proximity Analysis...")
        return {"next": "proximity"}

    elif "meta_review_text" not in state:
        print("[SUPERVISOR] Sending to Meta Review...")
        return {"next": "meta_review"}

    else:
        print("[SUPERVISOR] Sending to Summary...")
        return {"next": "summary"}


def generative_agent(state: dict) -> dict:
    query = state["query"]
    print(f"[GENERATION] Searching DuckDuckGo for: {query}")
    search_results = duck_search.run(query)
    summary = llm.invoke(f"Summarize the following search results:\n{search_results}")
    hypothesis = llm.invoke(f"Generate a hypothesis based on this summary:\n{summary}")
    print(f"[GENERATION] Hypothesis Generated: {hypothesis}")
    return {"hypothesis": hypothesis}



def reflection_agent(state: dict) -> dict:
    hypothesis = state["hypothesis"]
    print(f"[REFLECTION] Analyzing hypothesis: {hypothesis}")

    # Use LLM to generate feedback
    feedback = llm.invoke(f"Analyze this hypothesis and provide feedback on its coherence and completeness:\n{hypothesis}")

    print(f"[REFLECTION] Feedback Given: {feedback}")
    return {"feedback": feedback}

def ranking_agent(state: dict) -> dict:
    hypothesis = state["hypothesis"]
    feedback = state["feedback"]

    # Use LLM to generate a score based on hypothesis and feedback
    prompt = f"""
    Rate this hypothesis on a scale from 1-10 based on quality, accuracy, and relevance:

    Hypothesis: {hypothesis}
    Feedback: {feedback}

    Return only the numerical score.
    """
    score_text = llm.invoke(prompt)

    # Handle case where score_text is a tuple or not a string
    if isinstance(score_text, tuple):
        # Extract first element if it's a tuple
        score_text = str(score_text[0])
    else:
        score_text = str(score_text)

    # Extract digits and convert to integer
    digits = ''.join(c for c in score_text if c.isdigit())
    score = int(digits[:1]) if digits else 7  # Default to 7 if no digits found

    print(f"[RANKING] Score Assigned: {score}")
    return {"score": score}


def evolution_agent(state: dict) -> dict:
    hypothesis = state["hypothesis"]
    feedback = state["feedback"]
    score = state["score"]

    prompt = f"""
    Refine and improve this hypothesis based on the feedback and score:

    Original hypothesis: {hypothesis}
    Feedback: {feedback}
    Score: {score}/10

    Provide an improved version of the hypothesis.
    """

    refined_hypothesis = llm.invoke(prompt)

    print(f"[EVOLUTION] Updated Hypothesis: {refined_hypothesis}")
    return {"refined_hypothesis": refined_hypothesis}


def proximity_agent(state: dict) -> dict:
    # Should compare the refined hypothesis with previous similar hypotheses
    # or with a knowledge base to determine how novel or similar it is
    hypothesis = state["refined_hypothesis"]
    original_hypothesis = state["hypothesis"]

    # Using LLM to calculate similarity
    prompt = f"""
    Compare these two hypotheses and rate their similarity on a scale from 0.0 to 1.0:

    Original: {original_hypothesis}
    Refined: {hypothesis}

    Return only the numerical score between 0.0 and 1.0.
    """

    # In a real implementation, you might use embeddings or vector comparison
    # instead of asking the LLM to do this
    similarity_text = llm.invoke(prompt)

    # Parse the response to get a number
    try:
        # Extract all numbers from the text
        import re
        numbers = re.findall(r"0\.\d+|\d+\.\d+|\d+", str(similarity_text))
        if numbers:
            proximity_score = float(numbers[0])
            # Ensure it's in valid range
            proximity_score = max(0.0, min(1.0, proximity_score))
        else:
            proximity_score = 0.7  # Default if no number found
    except:
        proximity_score = 0.7  # Fallback value

    print(f"[PROXIMITY] Similarity Score: {proximity_score}")
    return {"proximity_score": proximity_score}

def meta_review_agent(state: dict) -> dict:
    # This should analyze the entire process and give feedback on how it went
    # It should look at all the previous steps

    # Gather all the relevant state information
    query = state["query"]
    hypothesis = state["hypothesis"]
    feedback = state["feedback"]
    score = state["score"]
    refined_hypothesis = state["refined_hypothesis"]
    proximity_score = state["proximity_score"]

    prompt = f"""
    Review this hypothesis generation process and provide constructive feedback:

    Query: {query}
    Initial Hypothesis: {hypothesis}
    Feedback Received: {feedback}
    Score Given: {score}/10
    Refined Hypothesis: {refined_hypothesis}
    Similarity Score: {proximity_score}

    What went well and what could be improved in this process? Focus on methodology.
    """

    review = llm.invoke(prompt)

    print(f"[META REVIEW] Review: {review}")
    return {"meta_review_text": review}
# ✅ Summary Agent
def summary_agent(state: dict) -> dict:
    summary = f"""
    Query: {state["query"]}
    Initial Hypothesis: {state["hypothesis"]}
    Reflection Feedback: {state["feedback"]}
    Initial Score: {state["score"]}
    Refined Hypothesis: {state["refined_hypothesis"]}
    Proximity Score: {state["proximity_score"]}
    Meta Review: {state["meta_review_text"]}
    """

    print("[FINAL SUMMARY]\n", summary)
    return {"summary": summary}


builder = StateGraph(LLMAgentCall)


builder.add_node("supervisor", supervisor_agent)
builder.add_node("generation", generative_agent)
builder.add_node("reflection", reflection_agent)
builder.add_node("ranking", ranking_agent)
builder.add_node("evolution", evolution_agent)
builder.add_node("proximity", proximity_agent)
builder.add_node("meta_review", meta_review_agent)
builder.add_node("summary", summary_agent)


builder.add_edge(START, "supervisor")

# Fix the conditional edges to properly route based on the "next" key
builder.add_conditional_edges(
    "supervisor",
    lambda state: state["next"],
    {
        "generation": "generation",
        "reflection": "reflection",
        "ranking": "ranking",
        "evolution": "evolution",
        "proximity": "proximity",
        "meta_review": "meta_review",
        "summary": "summary"
    }
)

builder.add_edge("generation", "supervisor")
builder.add_edge("reflection", "supervisor")
builder.add_edge("ranking", "supervisor")
builder.add_edge("evolution", "supervisor")
builder.add_edge("proximity", "supervisor")
builder.add_edge("meta_review", "supervisor")
builder.add_edge("summary", END)

# ✅ Compile Graph
graph = builder.compile()
```

```python
import redis
import hashlib
import json
from typing import Optional
import time
import logging

# Set up logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
```

```python
# Redis connection with retry logic
def get_redis_connection(max_retries=3, retry_delay=1):
    """Establish Redis connection with retry logic"""
    for attempt in range(max_retries):
        try:
            client = redis.Redis(
                host='localhost',
                port=6379,
                db=0,
                decode_responses=True,
                socket_connect_timeout=5
            )
            # Test the connection
            client.ping()
            logger.info("Successfully connected to Redis")
            return client
        except redis.ConnectionError as e:
            logger.warning(f"Redis connection attempt {attempt+1}/{max_retries} failed: {e}")
            if attempt < max_retries - 1:
                time.sleep(retry_delay)
            else:
                logger.error("Could not connect to Redis after maximum retries")
                return None

# Initialize Redis client with retry mechanism
redis_client = redis.Redis(host='127.0.0.1', port=6379, db=0, decode_responses=True)
CACHE_EXPIRATION = 60 * 60 * 24  # 24 hours in seconds
```

```python
import json

def cached_llm_invoke(prompt: str, model_name: Optional[str] = None, variation_level: float = 0.2) -> str:
    """
    Wrapper for LLM invocation that uses Redis for caching but still allows for response variation.
    """
    # Create a cache key from the prompt and model name
    cache_key = f"llm_cache:{hashlib.md5((prompt + (model_name or 'default')).encode()).hexdigest()}"

    # Check if response is in cache
    cached_response = redis_client.get(cache_key)
    if cached_response:
        print(f"[CACHE] Hit for prompt: {prompt[:50]}...")

        # Instead of directly returning the cached response, we use it as context for variation
        if variation_level > 0:
            variation_prompt = f"""
            I need a response to the following prompt:

            {prompt}

            Here is a previous response I've given that you can use as reference:

            {cached_response}

            Please provide a response that captures the same information but with {int(variation_level * 100)}% variation
            in wording, structure, or examples. Maintain the same level of accuracy and helpfulness.
            """

            # Get a varied response based on the cached one
            response = llm.invoke(variation_prompt)
            print(f"[CACHE] Generated variant with {int(variation_level * 100)}% variation level")
            return response
        else:
            # If variation_level is 0, just return the cached response directly
            return cached_response

    # If not in cache, call the LLM
    print(f"[CACHE] Miss for prompt: {prompt[:50]}...")
    response = llm.invoke(prompt)

    try:
        # Convert the AIMessage object to a string before storing in Redis
        if hasattr(response, 'content'):
            # For AIMessage objects that have a 'content' attribute
            redis_client.set(cache_key, str(response.content), ex=CACHE_EXPIRATION)
        elif isinstance(response, str):
            # For string responses
            redis_client.set(cache_key, response, ex=CACHE_EXPIRATION)
        else:
            # For other object types, use JSON serialization
            redis_client.set(cache_key, json.dumps(str(response)), ex=CACHE_EXPIRATION)
    except Exception as e:
        print(f"WARNING:__main__:Failed to write to cache: {e}")

    return response
```

```python
def ranking_agent(state: dict) -> dict:
    hypothesis = state["hypothesis"]
    feedback = state["feedback"]

    # Use cached LLM to generate a score
    prompt = f"""
    Rate this hypothesis on a scale from 1-10 based on quality, accuracy, and relevance:

    Hypothesis: {hypothesis}
    Feedback: {feedback}

    Return only the numerical score.
    """
    score_text = cached_llm_invoke(prompt)

    # Handle case where score_text is a tuple or not a string
    if isinstance(score_text, tuple):
        # Extract first element if it's a tuple
        score_text = str(score_text[0])
    else:
        score_text = str(score_text)

    # Extract digits and convert to integer
    digits = ''.join(c for c in score_text if c.isdigit())
    score = int(digits[:1]) if digits else 7  # Default to 7 if no digits found

    print(f"[RANKING] Score Assigned: {score}")
    return {"score": score}

```

```python
def evolution_agent(state: dict) -> dict:
    hypothesis = state["hypothesis"]
    feedback = state["feedback"]
    score = state["score"]

    prompt = f"""
    Refine and improve this hypothesis based on the feedback and score:

    Original hypothesis: {hypothesis}
    Feedback: {feedback}
    Score: {score}/10

    Provide an improved version of the hypothesis.
    """

    refined_hypothesis = cached_llm_invoke(prompt)

    print(f"[EVOLUTION] Updated Hypothesis: {refined_hypothesis}")
    return {"refined_hypothesis": refined_hypothesis}

```

```python
def proximity_agent(state: dict) -> dict:
    hypothesis = state["refined_hypothesis"]
    original_hypothesis = state["hypothesis"]

    # Using cached LLM to calculate similarity
    prompt = f"""
    Compare these two hypotheses and rate their similarity on a scale from 0.0 to 1.0:

    Original: {original_hypothesis}
    Refined: {hypothesis}

    Return only the numerical score between 0.0 and 1.0.
    """

    similarity_text = cached_llm_invoke(prompt)

    # Parse the response to get a number
    try:
        # Extract all numbers from the text
        numbers = re.findall(r"0\.\d+|\d+\.\d+|\d+", str(similarity_text))
        if numbers:
            proximity_score = float(numbers[0])
            # Ensure it's in valid range
            proximity_score = max(0.0, min(1.0, proximity_score))
        else:
            proximity_score = 0.7  # Default if no number found
    except:
        proximity_score = 0.7  # Fallback value

    print(f"[PROXIMITY] Similarity Score: {proximity_score}")
    return {"proximity_score": proximity_score}
```

```python
def meta_review_agent(state: dict) -> dict:
    # Gather all the relevant state information
    query = state["query"]
    hypothesis = state["hypothesis"]
    feedback = state["feedback"]
    score = state["score"]
    refined_hypothesis = state["refined_hypothesis"]
    proximity_score = state["proximity_score"]

    prompt = f"""
    Review this hypothesis generation process and provide constructive feedback:

    Query: {query}
    Initial Hypothesis: {hypothesis}
    Feedback Received: {feedback}
    Score Given: {score}/10
    Refined Hypothesis: {refined_hypothesis}
    Similarity Score: {proximity_score}

    What went well and what could be improved in this process? Focus on methodology.
    """

    review = cached_llm_invoke(prompt)

    print(f"[META REVIEW] Review: {review}")
    return {"meta_review_text": review}
```

```python
def summary_agent(state: dict) -> dict:
    summary = f"""
    Query: {state["query"]}
    Initial Hypothesis: {state["hypothesis"]}
    Reflection Feedback: {state["feedback"]}
    Initial Score: {state["score"]}
    Refined Hypothesis: {state["refined_hypothesis"]}
    Proximity Score: {state["proximity_score"]}
    Meta Review: {state["meta_review_text"]}
    """

    print("[FINAL SUMMARY]\n", summary)
    return {"summary": summary}
```

```python
def clear_cache():
    """Clear all cached LLM responses"""
    keys = redis_client.keys("llm_cache:*") + redis_client.keys("search_cache:*")
    if keys:
        redis_client.delete(*keys)
        print(f"Cleared {len(keys)} cached items")
    else:
        print("Cache already empty")
```

```python
builder = StateGraph(LLMAgentCall)

builder.add_node("supervisor", supervisor_agent)
builder.add_node("generation", generative_agent)
builder.add_node("reflection", reflection_agent)
builder.add_node("ranking", ranking_agent)
builder.add_node("evolution", evolution_agent)
builder.add_node("proximity", proximity_agent)
builder.add_node("meta_review", meta_review_agent)
builder.add_node("summary", summary_agent)

builder.add_edge(START, "supervisor")

# Fix the conditional edges to properly route based on the "next" key
builder.add_conditional_edges(
    "supervisor",
    lambda state: state["next"],
    {
        "generation": "generation",
        "reflection": "reflection",
        "ranking": "ranking",
        "evolution": "evolution",
        "proximity": "proximity",
        "meta_review": "meta_review",
        "summary": "summary"
    }
)

builder.add_edge("generation", "supervisor")
builder.add_edge("reflection", "supervisor")
builder.add_edge("ranking", "supervisor")
builder.add_edge("evolution", "supervisor")
builder.add_edge("proximity", "supervisor")
builder.add_edge("meta_review", "supervisor")
builder.add_edge("summary", END)

# Compile Graph
graph = builder.compile()
```

```python
from IPython.display import Image, display
import time
import requests

def display_graph_with_retry(graph, max_retries=3, retry_delay=2):
    """Displays the graph visualization with retry mechanism."""
    for attempt in range(max_retries):
        try:
            # Removed timeout argument as it's not supported
            display(Image(graph.get_graph().draw_mermaid_png()))
            return  # Exit if successful
        except requests.exceptions.ReadTimeout as e:
            print(f"Attempt {attempt + 1} failed: {e}")
            if attempt < max_retries - 1:
                print(f"Retrying in {retry_delay} seconds...")
                time.sleep(retry_delay)
            else:
                print("Max retries reached. Giving up.")

# Call the function without timeout argument:
display_graph_with_retry(graph, max_retries=3, retry_delay=10)
```

```python
def run_graph_with_cache_stats(query: str):
    """Run the graph and display cache statistics"""
    # Record initial cache size
    initial_cache_size = len(redis_client.keys("llm_cache:*"))

    # Run the graph
    result = graph.invoke({"query": query})

    # Record final cache size
    final_cache_size = len(redis_client.keys("llm_cache:*"))
    new_entries = final_cache_size - initial_cache_size

    print(f"\n[CACHE STATS] Initial cache entries: {initial_cache_size}")
    print(f"[CACHE STATS] New cache entries: {new_entries}")
    print(f"[CACHE STATS] Total cache entries: {final_cache_size}")

    return result
```

```python
def clean_text(text):
    # Replace escaped newlines and unnecessary spaces
    text = re.sub(r'\n+', '\n', text)
    text = re.sub(r'\s+', ' ', text)

    # Define sections for structured formatting
    sections = {
        "Query": r'🔍 **Query:**(.*?)🧠',
        "Initial Hypothesis": r'🧠 **Initial Hypothesis:**(.*?)💬',
        "Feedback": r'💬 **Feedback:**(.*?)📊',
        "Hypothesis Score": r'📊 **Hypothesis Score:**(.*?)🎯',
        "Refined Hypothesis": r'🎯 **Refined Hypothesis:**(.*?)📏',
        "Proximity Score": r'📏 **Proximity Score:**(.*?)📝',
        "Meta Review": r'📝 **Meta Review:**(.*?)➡️',
    }

    structured_output = {}

    for key, pattern in sections.items():
        match = re.search(pattern, text, re.DOTALL)
        structured_output[key] = match.group(1).strip() if match else ""

    return structured_output
```
